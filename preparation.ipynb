```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('educ_uoe_enrt01_linear.csv')
```

```python
df.dropna(subset=['OBS_VALUE'], inplace=True)
df = df.drop(columns = ['DATAFLOW', 'LAST UPDATE', 'freq', 'unit', 'OBS_FLAG'])
df.head()
```

```python
df = df[df['geo'] != 'EU28']
df = df[df['TIME_PERIOD'] != 2022]
df = df[df['TIME_PERIOD'] != 2012]
df = df[df['TIME_PERIOD'] != 2010]
df = df[df['TIME_PERIOD'] != 2005]
```

```python
df = df.sort_values(by = 'OBS_VALUE', ascending = False)
```

```python
countries_codes = pd.read_csv('eu_country_codes.csv', header = None)
```

```python
df = df.reset_index(drop=True)
countries_codes = countries_codes.reset_index(drop=True)
```

```python
countries_codes.head(5)
```

```python
countries_codes = countries_codes.rename(columns={0: 'geo', 1: 'name'})[['geo', 'name']]
```

```python
countries_codes
```

```python
df = df.merge(countries_codes, how='left', on = 'geo')
```

```python
df = df.drop(columns = ['geo'])
```

```python
df.to_csv('countries.csv', index = False)
```

```python
df = df[df['sex'] == 'T']
df = df[df['worktime'] == 'TOT_FTE']
df = df[df['sector'] == 'TOT_SEC']
df = df[df['TIME_PERIOD'] == 2021]
df = df[df['name'] == 'Malta']
```

```python
df = df[['isced11', 'OBS_VALUE']]
df
```

```python
full = df[df['isced11'] == 'ED5-8']['OBS_VALUE'].values[0]
full
```
